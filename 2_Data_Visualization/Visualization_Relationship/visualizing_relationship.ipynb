### Visualization Data Relationships:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the bee colony dataset
# Replace 'bee_colony_data.csv' with the actual path to your dataset
df = pd.read_csv('C:/Users/Maman Sauda/Desktop/ArewaDS-Machine-Learning-Assignments/2_Data_Visualization/Visualization_Relationship/Bee_data')

# Display the first few rows to understand the structure
df.head()
```

```python
# Visualize the health of the bee population state by state and year by year

plt.figure(figsize=(16, 8))
for state in df['state'].unique():
    state_data = df[df['state'] == state]
    plt.plot(state_data['year'], state_data['numcol'], label=state, alpha=0.5)
plt.title('Bee Colony Numbers by State and Year')
plt.xlabel('Year')
plt.ylabel('Number of Colonies')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
plt.tight_layout()
plt.show()
```

```python
# 1. Bar chart: Total honey production by state (sum over all years)
plt.figure(figsize=(14, 6))
state_prod = df.groupby('state')['totalprod'].sum().sort_values(ascending=False)
state_prod.plot(kind='bar')
plt.title('Total Honey Production by State (All Years)')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.tight_layout()
plt.show()
```

```python


# 3. Scatter plot: Number of colonies vs. yield per colony
plt.figure(figsize=(8, 6))
plt.scatter(df['numcol'], df['yieldpercol'], alpha=0.6)
plt.title('Number of Colonies vs. Yield per Colony')
plt.xlabel('Number of Colonies')
plt.ylabel('Yield per Colony')
plt.tight_layout()
plt.show()
```
