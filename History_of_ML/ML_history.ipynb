#### Timeline For Machine Learning


### Timeline: Geoffrey Hinton and the Rise of Deep Learning

**1970s**  
- Geoffrey Hinton begins research on neural networks and cognitive psychology.

**1986**  
- Hinton co-authors the landmark paper on backpropagation, revolutionizing neural network training.

**2006**  
- Hinton introduces "deep belief networks," sparking renewed interest in deep learning.

**2012**  
- Hinton and his students develop AlexNet, which wins the ImageNet competition and ignites the deep learning boom.

**2018**  
- Hinton receives the Turing Award for his contributions to deep learning.

**Present**  
- Hinton continues to advance AI research, focusing on capsule networks and unsupervised learning.
